```python
from graphviz import Digraph, Source
```

```python
class Dottable:
    def Dot(self):
        dot = Digraph()
        
        for i in range(self.N):
            dot.node(str(i))
            
        for i in range(self.N):
            if i != self.sets[i]:
                dot.edge(str(i), str(self.sets[i]))
                
        dot.graph_attr['rankdir'] = 'BT'
        return dot
```

```python
class DisjointSetQF(Dottable):
    def __init__(self, n):
        self.N = n
        self.sets = [i for i in range(n)]
        
    def Find(self, a):
        return self.sets[a]
    
    def Union(self, a, b):
        pa = self.Find(a)
        pb = self.Find(b)
        for i in range(self.N):
            if self.sets[i] == pa:
                self.sets[i] = pb
```

```python
ds = DisjointSetQF(10)
ds.Dot()
```

```python
ds.Union(3, 4)
ds.Dot()
```
